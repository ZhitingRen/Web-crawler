```python
from serpapi import GoogleSearch
import pandas as pd

# Your SerpAPI key
api_key = '8hshshs2384949447748494'  # Replace with your actual API key

def search_scholar(query, api_key, num_results=100):
    all_data = []
    results_per_page = 20  # Number of results per page (Google Scholar limit)
    for start in range(0, num_results, results_per_page):
        search_params = {
            'engine': 'google_scholar',
            'q': query,
            'api_key': api_key,
            'start': start,
            'num': results_per_page
        }

        search = GoogleSearch(search_params)
        results = search.get_dict()

        if 'error' in results:
            raise Exception(results['error'])

        publications = results.get('organic_results', [])

        for publication in publications:
            pub_info = publication.get('publication_info', {})
            authors = ', '.join(author['name'] for author in pub_info.get('authors', []))
            year = pub_info.get('year', '')

            all_data.append({
                'title': publication.get('title'),
                'author': authors,
                'abstract': publication.get('snippet', ''),
                'url': publication.get('link', ''),
                'year': year
            })

    return all_data

# Perform the search
query = 'online payment fraud'
data = search_scholar(query, api_key, num_results=100)  # Fetch 100 results

# Convert to pandas DataFrame
df = pd.DataFrame(data)

# Specify the file path to save the CSV file in D:\
file_path = 'D:\\scholarPaper_data.csv'

# Save to CSV
df.to_csv(file_path, index=False)

# Display the DataFrame
df.head()
```
